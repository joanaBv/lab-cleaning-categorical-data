### Abbridged version of lab-cleaning-numerical-data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
customer_df=pd.read_csv("files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
# converting column names to snake type

customer_df.columns = [x.lower() for x in customer_df.columns]
customer_df.columns = customer_df.columns.str.replace(' ','_')
display(customer_df.head())
```

```python
# turning effective to date into datetime type

customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
```

```python
# checking for nulls

customer_df.isnull().sum() # - no nulls
```

```python
# Splitting the Dataframe

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

numericals = customer_df.select_dtypes(include=numerics)

categoricals = customer_df.select_dtypes(include=object)
```

```python
def cont_disc(numericals):
    
    cont = pd.DataFrame()
    disc = pd.DataFrame()
    notdisc = cont.columns
    
    for col in numericals.columns:
        if numericals[col].dtypes == float:
            cont = pd.concat([cont,numericals[col]], axis=1)
     
    disc = numericals.drop(notdisc, axis=1)     
    
    return (cont, disc)   
```

```python
[continuous_df ,discrete_df ] = cont_disc(numericals)

display(continuous_df.head())
display(discrete_df .head())
```

```python
def uvalue_check(categoricals):
    
    uniques = {col: categoricals[col].unique() for col in categoricals.columns}
    
    return  (uniques)
```

```python
uvalue_check(categoricals)

# There are no nan values and nothing apparent that needs correction or deletion
```

```python
# The customer_id is a code unique to each customer, which would only add noise and no useful info to our analysis.
# Therefore the column should be dropped and disregarded - does not matter which customer is which.

categoricals = categoricals.drop('customer',axis=1)

categoricals.head()
```

## Start of lab-cleaning-categorical-data

```python
correlations_matrix = numericals.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()

# highest correlation is between monthly premium auto and total claim amount - which makes sense, the bigger your claim
# more expensive the monthly payments are
# followed by monthly premium auto and customer lifetime value - which also makes sense, the more you pay monthly the more
# you've paid overall minus the total claim associated to your policy 
```

```python
# Before plotting my numericals I'm going to normalize them so I can use my functions more effectively

# STANDARDSCALER

def stdnorm (df):
    
    standardnorm = StandardScaler().fit(df)
    x_st = standardnorm.transform(df)
    std_df = pd.DataFrame(x_st, columns=df.columns)
    
    return (std_df)
```

```python
std_continuous_df = stdnorm(continuous_df)
std_discrete_df = stdnorm(discrete_df)

display(std_continuous_df)
display(std_discrete_df)
```

```python
def disc_plot (disc):
    
    for col in disc.columns:
        sns.displot(disc[col], bins=20)
        plt.show()
```

```python
def cont_plot (cont):
    
    for col in cont.columns:
        plt.xlabel(col)
        plt.ylabel('Count')
        sns.kdeplot(cont[col],shade=True)
        plt.show()  
```

```python
disc_plot(std_discrete_df)
```

```python
cont_plot(std_continuous_df)
```

```python
cont_boxplot = std_continuous_df.boxplot()

# As confirmed by the boxplot below, 75% of all values of both the columns of continuous values dataframe lie below 2
# but we can see a steady stream of datapoints up to 6.
# I would classify the 4 max most datapoints from total claim amount, and the 3 max most datapoints from customer lifetime
# value, as outliers.
```
